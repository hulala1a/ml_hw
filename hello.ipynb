```python
import pandas as pd
import numpy as np
data = pd.read_csv('covid.test.csv')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

fig,axs = plt.subplots(ncols=5, figsize=(30,5))
sns.pointplot(x="wearing_mask", y="tested_positive", data=data, ax=axs[0])
sns.pointplot(x="depressed", y="tested_positive",data=data, ax=axs[1])
sns.pointplot(x="anxious", y="tested_positive", data=data, ax=axs[2])
```
